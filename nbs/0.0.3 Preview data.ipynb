```python
import sqlalchemy as sa
import pandas as pd
```

```python
def read_sql_data_frame(path):
    engine = sa.create_engine("sqlite:///%s" % path)
    connection = engine.connect()

    return pd.read_sql("experiments", connection)
```

```python
df = read_sql_data_frame('../data/without_gambler/the_main.db')
```

```python
df.describe()
```

```python
min(df.exp_index), max(df.exp_index)
```

```python
other_df = read_sql_data_frame('../data/without_gambler/main.db')
```

```python
min(other_df.exp_index), max(other_df.exp_index)
```

```python
data = pd.concat([df, other_df])
data = data.sort_values(by=['exp_index'])
data = data.reset_index(drop=True)
data = data.drop_duplicates()
```

```python
data.to_csv('../data/numerical_experiments_without_gambler.csv')
```
