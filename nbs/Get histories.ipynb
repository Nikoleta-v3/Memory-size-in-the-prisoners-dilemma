```python
import opt_mo
import random
```

```python
engine = sa.create_engine('sqlite:///../data/without_gambler/main.db')
```

```python
connection = engine.connect()
```

```python
df = pd.read_sql('experiments', connection)
```

```python
def get_histories(seed, function):
    axl.seed(seed)
    opponents = [[random.random() for _ in range(4)] for _ in range(2)]
    
    _, history, _ = opt_mo.get_evolutionary_best_response(opponents=opponents,
                                                          best_response_function=function)
    
    return history
```

**Creating histories**

```python
mem_histories = []
for seed in seeds:
    mem_histories.append(get_histories(seed, opt_mo.get_memory_one_best_response))    
```

```python
histories = []
for seed in reactive_seeds:
    histories.append(get_histories(seed, opt_mo.get_reactive_best_response_with_bayesian))    
```

```python
opt_mo.evolutionary_best_response.get_repeat_length_in_history(mem_histories[6])
```

```python
columns = [['seed: {} p: {}'.format(seed, prop) for prop in range(1, 5)] for seed in seeds]
columns = [item for sublist in columns for item in sublist]
```

```python
count = 0
series = []
for i, column in enumerate(columns):
    if i % 4 == 0:
        count += 1

    series.append(pd.Series([item[i % 4] for item in mem_histories[count -1]]))
```

```python
df_mem_history = pd.concat(series, ignore_index=True, axis=1)
df_mem_history.columns = columns
```
