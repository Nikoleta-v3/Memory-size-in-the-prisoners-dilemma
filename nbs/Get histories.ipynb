```python
def check_out_put(seed):
    axl.seed(seed)
    opponents = [[random.random() for _ in range(4)] for _ in range(2)]

    row = experiments_with_big_cycle[
        experiments_with_big_cycle["exp_index"] == 8
    ].values[0]

    assert np.allclose(row[1:5], opponents[0])
    assert np.allclose(row[5:9], opponents[1])

    best_response = opt_mo.get_memory_one_best_response(opponents)
    assert np.allclose(row[9:13], best_response)
    best_evol_response = opt_mo.get_evolutionary_best_response(
        opponents, opt_mo.get_memory_one_best_response
    )
    assert (
        opt_mo.evolutionary_best_response.get_repeat_length_in_history(
            best_evol_response[-2]
        )
        == best_evol_response[-1]
    )
    assert np.allclose(best_evol_response[0], row[13:17])
```

```python
check_out_put(8)
```
